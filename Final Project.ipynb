```python
import warnings; warnings.simplefilter("ignore")
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
information_columns = pd.ExcelFile("CHSI DataSet.xls")
dataframe1 = pd.read_csv("RISKFACTORSANDACCESSTOCARE.csv")
dataframe2 = pd.read_csv("DEMOGRAPHICS.csv")
dataframe3 = pd.read_csv("MEASURESOFBIRTHANDDEATH.csv")
```

```python
df_demo = dataframe2[["CHSI_County_Name","CHSI_State_Name","Poverty","Max_Poverty","Min_Poverty","Population_Size","County_FIPS_Code","State_FIPS_Code"]]
df_demo.head(10)
```

```python
df_riskfactor = dataframe1[["State_FIPS_Code","County_FIPS_Code","CHSI_County_Name","CHSI_State_Name","No_Exercise","Obesity","High_Blood_Pres","Smoker","Diabetes","Uninsured","Elderly_Medicare","Disabled_Medicare","Prim_Care_Phys_Rate"]]
```

```python
df_riskfactor.head(10)
```

```python
df_measurebd = dataframe3[["State_FIPS_Code","County_FIPS_Code","CHSI_County_Name","CHSI_State_Name","Late_Care","Infant_Mortality","Total_Deaths","Total_Births"]]
df_measurebd.head(10)
```

```python
df_demo_risk = pd.merge(df_demo, df_riskfactor, on=['CHSI_State_Name','CHSI_County_Name'])
df_demo_risk.head(10)
```

```python
df_demo_risk_bd = pd.merge(df_demo_risk, df_measurebd, on=['CHSI_State_Name','CHSI_County_Name'])
df_demo_risk_bd.head(10)
```

```python
df_final_1 = df_demo_risk_bd.groupby(['CHSI_State_Name','CHSI_County_Name',"State_FIPS_Code","State_FIPS_Code"])["No_Exercise","Obesity","Poverty","High_Blood_Pres","Smoker","Diabetes","Late_Care","Infant_Mortality","Total_Deaths","Total_Births"].sum()
df_final_1 = df_final_1.replace(-1111,0)
df_final_1 = df_final_1.replace(-1111.1,0)
df_final_1 = df_final_1.replace(-1, 0)
df_final_1 = df_final_1.replace(-2222.2, 0)
df_final_1 = df_final_1.replace(-2222, 0)
df_final_1 = df_final_1.replace(-2, 0)
df_final_1

```

```python
df_final_2 = df_demo_risk_bd.groupby(['CHSI_State_Name','CHSI_County_Name',"State_FIPS_Code","County_FIPS_Code"])["Uninsured","Elderly_Medicare","Disabled_Medicare","Prim_Care_Phys_Rate","Poverty","Late_Care","Infant_Mortality","Total_Deaths","Total_Births"].sum()
df_final_2 = df_final_2.replace(-1111,0)
df_final_2 = df_final_2.replace(-1111.1,0)
df_final_2 = df_final_2.replace(-1, 0)
df_final_2 = df_final_2.replace(-2222.2, 0)
df_final_2 = df_final_2.replace(-2222, 0)
df_final_2 = df_final_2.replace(-2, 0)
df_final_2
```
